```python
from google_play_scraper import app, reviews, Sort, reviews_all # Scraper from google play
import csv
import requests
from io import StringIO
import pandas as pd
```

## **2. Data Scraping**

```python
scrapreview = reviews(
    'notion.id',          # App ID
    lang='en',
    country='us',
    sort=Sort.MOST_RELEVANT,
    count = 15000 # Numbers of review to scrape
)
```

```python
app_reviews_df = pd.DataFrame(scrapreview[0])
jumlah_ulasan, jumlah_kolom = app_reviews_df.shape
```

## **3. Data Preprocessing**

### Drop Irrelevant Features

```python
clean_df = app_reviews_df.drop(['reviewId', 'userName', 'userImage', 'at', 'replyContent', 'repliedAt', 'reviewCreatedVersion', 'appVersion'], axis=1)
clean_df = clean_df.dropna()
clean_df.to_csv('clean_data.csv', index=False)
clean_df.isnull().sum()
```
